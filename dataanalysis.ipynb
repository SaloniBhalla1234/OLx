```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df =pd.read_csv('olxcleaned.csv')
```

```python
lb = LabelEncoder()
df['Make_Encode'] = lb.fit_transform(df['Make'])
```

```python
lb = LabelEncoder()
feature = {} 
for k in df.columns:
     if df[k].dtypes=='O':
         df[str(k)+'_Encode'] = lb.fit_transform(df[k])
         gt = {}
         ke = df[[k,str(k)+'_Encode']].drop_duplicates()[str(k)+'_Encode'].values
         val = df[[k,str(k)+'_Encode']].drop_duplicates()[k].values
         for i in range(len(ke)):
             gt[ke[i]] = val[i]
         feature[k] = gt
```
